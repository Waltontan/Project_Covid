```python
import requests
from pprint import pprint
from config import nyt_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
```

```python
# Search for articles that mention granola
query = "covid"
begin_date = 20211029
end_date = 20211030
```

```python
# Build query URL
# query_url = url + "api-key=" + nyt_key + "&q=" + query + "&begin_date=" + begin_date + "&end_date=" + end_date
query_url = f'{url}api-key={nyt_key}&q={query}&begin_date={begin_date}&end_date={end_date}'
```

```python
# Request articles
articles = requests.get(query_url).json()
articles
```
