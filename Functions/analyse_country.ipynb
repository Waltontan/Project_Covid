```python
def analyse_country(df,col1,col2):
    import pandas as pd
    import scipy.stats as sts
    
    countries = df['location'].unique()
    r_squared = []
    for country in countries:
        temp_df = df.loc[df['location']==country]
        temp_df = temp_df [[col1,col2]]
        temp_df = temp_df.dropna()
        
        # Perform a linear regression on x_value vs. y_value
        (slope, intercept, rvalue, pvalue, stderr) = sts.linregress(temp_df[col1], temp_df[col2])        
        
        r_squared.append(rvalue**2)
        
    analyse_df = pd.DataFrame({'countries':countries,'R_Squared':r_squared})
    analyse_df = analyse_df.sort_values(by='R_Squared',ascending = False)
    
    
    return analyse_df
```
